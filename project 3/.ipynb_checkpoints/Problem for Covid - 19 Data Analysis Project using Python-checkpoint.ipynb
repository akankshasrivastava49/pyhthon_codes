# 5. Data Cleaning
        

#  a. Remove all duplicates observations

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

#    b. Find missing values in all columns

```python
data.isnull().sum()
```

#  c. Remove all observations where continent column value is missing

```python
data.dropna(subset=['continent'])
```

# d. Fill all missing values with 0

```python
data.isnull().sum()
```

```python
data['iso_code'] = data['iso_code'].fillna(data['iso_code'].mode()[0])
```

```python
data['continent'] = data['continent'].fillna(data['continent'].mode()[0])
```

```python
data['location'] = data['location'].fillna(data['location'].mode()[0])
```

```python
data['total_cases'] = data['total_cases'].fillna(data['total_cases'].mean())
```

```python
data['new_cases'] = data['new_cases'].fillna(data['new_cases'].mean())
```

```python
data['new_cases_smoothed'] = data['new_cases_smoothed'].fillna(data['new_cases_smoothed'].mean())
```

```python
data['total_deaths'] = data['total_deaths'].fillna(data['total_deaths'].mean())
```

```python
data['new_deaths'] = data['new_deaths'].fillna(data['new_deaths'].mean())
```

```python
data['new_deaths_smoothed'] = data['new_deaths_smoothed'].fillna(data['new_deaths_smoothed'].mean())
```

```python
data['total_cases_per_million'] = data['total_cases_per_million'].fillna(data['total_cases_per_million'].mean())
```

```python
data['new_cases_per_million'] = data['new_cases_per_million'].fillna(data['new_cases_per_million'].mean())
```

```python
data['new_cases_smoothed_per_million'] = data['new_cases_smoothed_per_million'].fillna(data['new_cases_smoothed_per_million'].mean())
```

```python
data['total_deaths_per_million'] = data['total_deaths_per_million'].fillna(data['total_deaths_per_million'].mean())
```

```python
data['new_deaths_per_million'] = data['new_deaths_per_million'].fillna(data['new_deaths_per_million'].mean())
```

```python
data['reproduction_rate']=data['reproduction_rate'].fillna(data['reproduction_rate'].mean())
```

```python
data.isnull().sum()
```

# 6. Date time format :

#  a. Convert date column in datetime format using pandas.to_datetime

```python
data.dtypes
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data.dtypes
```

# b. Create new column month after extracting month data from date column
        

```python
data['month'] = data['date'].dt.month
```

# 8. Feature Engineering :

#  a. Create a new feature 'total_deaths_to_total_cases' by ratio of
        'total_deaths' column to 'total_cases'
        

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```
